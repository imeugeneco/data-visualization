# Experimenting with Seaborn Displot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Generating 150 random numbers
varx = np.random.randn(150)
print(varx.size)
varx
```

```python
sns.set()
sns.distplot(varx, color='red')
```

```python
label_dist = pd.Series(varx, name='variable x')
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist, vertical=True, color='green', hist=False)
```

```python
sns.distplot(label_dist, color='gray', rug=True)
```

```python
# Loading the flights dataset
flights = sns.load_dataset('flights')
flights.head()
```

```python
# Plotting number of passengers
# rug=True shows the distribution
sns.distplot(flights['passengers'], rug=True, color='purple')
```

```python
# Feature engineering the number of passengers...
flights['passengers'].describe()
```

```python
for index, row in flights.iterrows():
    if row.passengers >= 500:
        flights.loc[index, 'passengers_fe'] = '500+'
    elif row.passengers >= 400:
        flights.loc[index, 'passengers_fe'] = '400-499'
    elif row.passengers >= 300:
        flights.loc[index, 'passengers_fe'] = '300-399'
    elif row.passengers >= 200:
        flights.loc[index, 'passengers_fe'] = '200-299'
    else:
        flights.loc[index, 'passengers_fe'] = 'below 200'

flights.passengers_fe.value_counts()
```
