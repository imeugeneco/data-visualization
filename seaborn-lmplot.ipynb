# Experimenting with Seaborn Lmplot

```python
import seaborn as sns
```

```python
# Loading the penguins dataset
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# Plotting the relevance between flipper length and body mass, by species
sns.lmplot(data=penguins, x='flipper_length_mm', y='body_mass_g', col='species', hue='species')
```

```python
# Strong link between flipper length and body mass, especially prevalent in the 'Chinstrap' and Gentoo' spcies
# Difference between species is also clear

# Plotting the relevance between culmen length and culmen depth, by species
sns.lmplot(data=penguins, x='culmen_length_mm', y='culmen_depth_mm', col='species', hue='species')
```

```python
# Adelie: short & deep culmen
# Chinstrap: long & deep culmen
# Gentoo: average length & shallow culmen

# Are the penguins' culmens favored by their habitat?
sns.lmplot(data=penguins, x='culmen_length_mm', y='culmen_depth_mm', hue='island', fit_reg=False)
```

```python
# Re-plotting the graph above...
sns.lmplot(data=penguins, x='culmen_length_mm', y='culmen_depth_mm', col='island', hue='species', fit_reg=False)
```

```python
# 'Adelie' species (short & deep culmen) live in all three islands
# 'Chinstrap' species (long & deep culmen) live in Dream island only
# 'Gentoo' species (average length & shallow culmen) live in Biscoe island only

sns.lmplot(data=penguins, x='culmen_length_mm', y='culmen_depth_mm', col='island', hue='island')
```

```python
# Is sex a factor in determining the penguins' culmen length & depth?
sns.lmplot(data=penguins, x='culmen_length_mm', y='culmen_depth_mm', hue='sex', fit_reg=False)
```

```python
# Is sex a factor in determining the flipper length or body mass?
sns.lmplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='sex', markers=['o', 'x'], fit_reg=False)
```

```python
# The above graph seems like the gap may be caused by the type of species involved...

sns.lmplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='sex', markers=['o', 'x'], fit_reg=False, col='species')
```

```python
# So in average, male penguins (regardless of species) is heavier in body mass & 
# has longer flippers than their female counterparts
```
